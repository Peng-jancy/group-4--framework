```python
import sys
sys.path.append("./model_one-Copy1")

from data.dataset import load_mnist
from models.cnn_model import ConvMyFlow
from utils.plot import plot_final_model

TOTAL_SAMPLES=2000
TEST_SAMPLE_NUM=200
train_epochs=200

X_train,y_train,X_test,y_test=load_mnist(TOTAL_SAMPLES,TEST_SAMPLE_NUM)

conv_config = [
{'type':'conv','out_channels':16,'kernel_size':3,'stride':1,'padding':1,'af':1},
{'type':'pool','pool_size':2,'stride':2},
{'type':'conv','out_channels':32,'kernel_size':3,'stride':1,'padding':1,'af':1},
{'type':'pool','pool_size':2,'stride':2}
]

hidden_layers=[128,64]

activation_list=[1,1,4]

model=ConvMyFlow(
conv_config,
hidden_layers,
X_train,
y_train,
10,
activation_list,
train_epochs,
0.001,
0.001
)

loss,train_acc,test_acc=model.train(X_test,y_test)

plot_final_model(loss,train_acc,test_acc,train_epochs)
```
